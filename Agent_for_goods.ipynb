```python
from google.colab import userdata
from google import genai

api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

print("Connected successfully!")
```

```python
from google.genai import types

def researcher_agent(topic):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=f"Research this topic thoroughly for a student report: {topic}. Find key facts, recent developments, and important context.",
        config=types.GenerateContentConfig(
            tools=[types.Tool(google_search=types.GoogleSearch())]
        )
    )
    return response.text

print("Researcher agent ready!")
```

```python
def analyzer_agent(research_text):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=f"You are an analyst. Extract the most important facts, statistics, and insights from this research. Organize them into clear bullet points grouped by theme:\n\n{research_text}"
    )
    return response.text

print("Analyzer agent ready!")
```

```python
def critic_agent(analysis_text, topic):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=f"You are a strict editor. Review this analysis on '{topic}' for gaps, bias, missing perspectives, or weak evidence. List specific issues to fix:\n\n{analysis_text}"
    )
    return response.text

print("Critic agent ready!")
```

```python
def report_writer_agent(topic, analysis_text, critic_feedback):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=f"""Write a well-structured research report on '{topic}'.

Use this analysis:
{analysis_text}

Address these critic notes:
{critic_feedback}

Format with a title, intro, clear sections with headers, and a conclusion. Make it professional and easy to read."""
    )
    return response.text

print("Report writer agent ready!")
```

```python
def run_research_pipeline(topic):
    print(f"🔍 Researching: {topic}...")
    research = researcher_agent(topic)

    print("🧠 Analyzing...")
    analysis = analyzer_agent(research)

    print("⚖️ Critiquing...")
    feedback = critic_agent(analysis, topic)

    print("📝 Writing final report...")
    report = report_writer_agent(topic, analysis, feedback)

    print("✅ Done!\n")
    return report

print("Pipeline ready!")
```

```python
topic = "How AI agents can improve access to rural healthcare"

final_report
final_report = run_research_pipeline(topic)
print(final_report)
```

```python
def save_report(topic, report_text):
    filename = topic.lower().replace(" ", "_")[:40] + "_report.md"
    with open(filename, "w") as f:
        f.write(report_text)
    print(f"Saved as {filename}")

save_report(topic, final_report)
```
